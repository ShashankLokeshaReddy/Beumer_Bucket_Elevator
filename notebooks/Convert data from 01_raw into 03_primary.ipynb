```python
def getdf(folder, save_to, file_name):
    df = pd.concat((pd.read_csv(f) for f in iglob(folder, recursive=True)), ignore_index=True)
    #print(df)
    #print(df[['Index', 'Range mm', 'i[1]', 'q[1]']])
    data = df.drop(df.columns.values[-1], axis=1)
    for i in range(len(data)):
        if pd.isnull(data.iat[i, len(data.columns)-1]) is True:
            data.iloc[i]= data.iloc[i].shift(periods=1,axis=0, fill_value=0)

    offset = 3
    new_columns = len(data.columns) - offset
    new_data = data.iloc[0:len(data), offset:len(data.columns)]
    new_data.insert(loc=0, column="Range mm", value=data["Index"])

    ivalues = []
    qvalues = []
    for i in range(int(new_columns/2.0)):
        ivalues.append(str("i["+str(i)+"]"))
        qvalues.append(str("q["+str(i)+"]"))

    amplitudes = []

    for ind in range(new_data.index.values.min(), new_data.index.values.max()+1):
        k = []
        for i in range(len(ivalues)):
            k.append(np.sqrt(new_data[new_data.index==ind][ivalues[i]].values[0]**2+new_data[new_data.index==ind][qvalues[i]].values[0]**2))
        amplitudes.append(k)
    #print(amplitudes) # calculatesamplitude from the i-q pairs.
    new_data = pd.DataFrame(data = amplitudes, columns = np.linspace(20, FSR, len(ivalues)))
    new_data.T
    new_data.to_csv(save_to + file_name)
    return new_data
```
